```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_columns = ['batsman', 'bowler', 'venue']
le = LabelEncoder()

for col in le_columns:
    df[col] = le.fit_transform(df[col])
```

```python
cat_cols = ['bat_team', 'bowl_team']
df = pd.concat((df, pd.get_dummies(df[cat_cols])), axis = 1)
```

```python
df
```

```python
df.drop(cat_cols, axis = 1, inplace = True)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['date'].dt.year
```

Splitting data into train and test sets: 

data before 2015 taken as train set to train model and use the model to predict scores after 2015 which can be checked with the actual values.

```python
train_set = df[df['date'].dt.year <= 2015]
test_set = df[df['date'].dt.year > 2015]

print('Shape of training set is %d, %d'%(train_set.shape[0], train_set.shape[1]))
print('Shape of testing set is %d, %d'%(test_set.shape[0], test_set.shape[1]))
```

```python
train_set.drop(['date'], axis = 1, inplace = True)
test_set.drop(['date'], axis = 1, inplace = True)
```

```python
from sklearn.preprocessing import StandardScaler

for column in train_set.columns:
  
  if column != 'total':
    ss = StandardScaler()
    train_set[column] = ss.fit_transform(train_set[column].values.reshape((-1,1)))

for column in test_set.columns:

  if column != 'total':
    ss = StandardScaler()
    test_set[column] = ss.fit_transform(test_set[column].values.reshape((-1,1)))
```

```python
x_train, y_train = train_set.drop('total', axis = 1), train_set['total']
x_test, y_test = test_set.drop('total', axis = 1), test_set['total']
```

```python
import xgboost
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

XGBoost is used here, because it is capable of capturing non-linear relationships, built-in mechanisms to handle missing data, enabling it to handle such scenarios without requiring extensive data preprocessing. It is very efficient and scalable. XGBoost's parallel processing capabilities and algorithmic optimizations contribute to its fast training and prediction times. XGBoost offers numerous hyperparameters that can be fine-tuned to optimize the model's performance for IPL cricket dataset prediction.

Creating model and fitting training data

XGBoost is an ensemble learning method that combines multiple decision trees to create a strong predictive model. It includes regularization techniques such as L1 and L2 regularization to prevent overfitting. This helps to improve the model's generalization.

```python
model = XGBRegressor()
model.fit(x_train, y_train)
```

Using model to predict scores

```python
y_pred = model.predict(x_test)
y_pred
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print("Mean absolute error: ", mae)
print("Mean squared error: ", mse)
```
